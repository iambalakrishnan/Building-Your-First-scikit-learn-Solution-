```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
housing_data = pd.read_csv('housing.csv')
housing_data.head()
```

```python
#lets look into random sample data

housing_data.sample(5)
```

```python
housing_data.shape
```

```python
#drop all the missing value
housing_data = housing_data.dropna()
```

```python
housing_data.shape
```

```python
housing_data.describe()
```

```python
#Lets look at catagorical column

housing_data['ocean_proximity'].unique()
```

```python
#Lets explore data set

fig, ax = plt.subplots(figsize=(12,8))

plt.scatter(housing_data['total_rooms'], housing_data['median_house_value'])

plt.title('Exploring The California Housing Datasets')
plt.xlabel('Total Rooms')
plt.ylabel('Median House Value')

plt.show()
```

```python
#Lets plot relationship between housing median age and median house value

fig, ax = plt.subplots(figsize=(12,8))

plt.scatter(housing_data['housing_median_age'], housing_data['median_house_value'])

plt.xlabel('Median Age')
plt.ylabel('Median House Value')

plt.show()
```

```python
#Lets comparw with median income

fig, ax = plt.subplots(figsize=(12,8))

#multiplied median income *10000 so it can show actual income
plt.scatter(housing_data['median_income'] * 10000, housing_data['median_house_value'])

plt.xlabel('Median Income')
plt.ylabel('Median House Value')

plt.show()
```

```python
#Lets find correlation

housing_data_corr = housing_data.corr()
housing_data_corr
```

median_income has slightly better correlation compared to total_rooms, median income is a better predicted of house prices compared with total_rooms<br>

```python
#Rather than viewing raw correlation numbers, correlations are best visualized using something called a heat map.

fig, ax = plt.subplots(figsize=(12,8))

sns.heatmap(housing_data_corr, annot=True)#annot=True will printout correlation values on our heatmap
#Darker color negative correlation and lighter color indicates positive corelation 
#Lets check which column has better positive correlatin with our target column
```
